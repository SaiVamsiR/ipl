```python
import pandas as pd
import numpy as np
```

```python
match=pd.read_csv('Matches_Result.csv')
```

```python
ball=pd.read_csv('Ball_by_Ball.csv')
```

```python
match.info()
```

```python
ball.head()
```

```python
ball=ball.merge(match[['ID','Team1','Team2','WinningTeam']],how='left')
```

```python
total_score=ball.groupby(['ID','innings']).sum()['total_run'].reset_index()
```

```python
total_score
```

```python
total_score=total_score[total_score['innings']==1]
```

```python
match_df=match.merge(total_score[['ID','total_run']],left_on='ID',right_on='ID')
```

```python
match_df.head()
```

```python
match_df['Team1'].unique()
```

```python
teams=['Rajasthan Royals', 'Royal Challengers Bangalore',
       'Sunrisers Hyderabad', 'Delhi Capitals', 'Chennai Super Kings',
       'Gujarat Titans', 'Lucknow Super Giants', 'Kolkata Knight Riders',
       'Punjab Kings', 'Mumbai Indians', 'Kings XI Punjab']
```

```python
match_df['Team1']=match_df['Team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['Team2']=match_df['Team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['Team1']=match_df['Team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['Team2']=match_df['Team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

match_df['Team1']=match_df['Team1'].str.replace('Gujarat Lions','Gujarat Titans')
match_df['Team2']=match_df['Team2'].str.replace('Gujarat Lions','Gujarat Titans')
```

```python
match_df=match_df[match_df['Team1'].isin(teams)]
match_df=match_df[match_df['Team2'].isin(teams)]
```

```python
match_df.columns
```

```python
match_df=match_df[['ID','City','total_run']]
```

```python
d_df=match_df.merge(ball,on='ID')
```

```python
d_df=d_df[d_df['innings']==2]
```

```python
d_df.shape
```

```python
d_df.head()
```

```python
# Convert to numeric (non-convertible values will be NaN)
d_df['total_run_y'] = pd.to_numeric(d_df['total_run_y'], errors='coerce')

# Optionally handle NaNs
d_df['total_run_y'] = d_df['total_run_y'].fillna(0)

# Now you can safely compute the cumulative sum
d_df['current_score'] = d_df.groupby('ID')['total_run_y'].cumsum()
```

```python
d_df.head(25)
```

```python
d_df['runs_left']=d_df['total_run_x']-d_df['current_score'] + 1
```

```python
d_df['ballnumber'].value_counts()
```

```python
d_df['balls_left']=120-(d_df['overs']*6+d_df['ballnumber'])
```

```python
# Convert to numeric: True/False will become 1/0, respectively
d_df['isWicketDelivery'] = d_df['isWicketDelivery'].astype(int)

# Now you can safely perform cumulative sum
d_df['wickets'] = 10 - d_df.groupby('ID')['isWicketDelivery'].cumsum()
```

```python
d_df['crr']=(d_df['current_score']*6)/(120-d_df['balls_left'])
```

```python
d_df['rrr']=(d_df['runs_left']*6)/d_df['balls_left'] 
```

```python
d_df.columns
```

```python
d_df.tail()
```

```python
d_df['BowlingTeam'] = d_df.apply(
    lambda row: row['Team2'] if row['BattingTeam'] == row['Team1'] else row['Team1'],
    axis=1
)
```

```python
def result(row):
    return 1 if row['BattingTeam']==row['WinningTeam'] else 0
```

```python
d_df['result']=d_df.apply(result,axis=1)
```

```python
d_df.columns
```

```python
final=d_df[['BattingTeam','BowlingTeam','City','runs_left','balls_left','wickets','total_run_x','crr','rrr','result']]
```

```python
final=final.sample(final.shape[0])
```

```python
final.isnull().sum()
```

```python
final.dropna(inplace=True)
```

```python
final=final[final['balls_left']!=0]
```

```python
X=final.iloc[:,:-1]
y=final.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
ct = ColumnTransformer([
    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False), ['BattingTeam', 'BowlingTeam', 'City']) # OneHotEncoder() is an instantiated object
], remainder='passthrough')
```

```python
pipe=Pipeline([
    ('ct',ct),
    ('reg',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
import pickle
pickle.dump(pipe,open('pipe.pkl','wb'))
```
